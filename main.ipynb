```python
characters = {
    "Researcher": "You are a researcher specialized in climate change. You should explain how global warming is affecting the Earth's temperature, weather patterns, and ecosystems based on scientific data and evidence. Additionally, compare the effects of climate change in various regions of the world, especially in both developed and developing countries.",
    "Philosopher": "You are a philosopher who analyzes the ethical and philosophical implications of global warming. Your task is to evaluate whether humans have a moral responsibility to address this crisis and how it challenges our values and philosophy towards nature.",
    "Artist": "You are an artist who uses metaphors, visuals, and emotions to express the impact of global warming. Your analysis should focus on how art can depict the emotional and visual aspects of the climate crisis, inspiring reflection and action.",
    "Politician": "You are a politician tasked with finding practical solutions to combat global warming. You should propose policies that balance economic, social, and environmental interests, aiming for global cooperation and national action."
}
```

```python
conversation_history = [
    {"role": "system", "content": "This is a discussion about global warming between a Researcher, a Philosopher, an Artist, and a Politician. Each participant responds directly to the previous messages, ensuring a logical and engaging conversation."},
    {"role": "user", "content": "Moderator: Let's discuss global warming. What are the biggest challenges we face?"}
]
```

```python
def generate_response(name):
    recent_history = conversation_history[-5:]  
    messages = recent_history + [{"role": "system", "content": characters[name]}]

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
    )
    return response.choices[0].message.content
```

```python
def simulate_conversation(rounds=5):
    for _ in range(rounds):
        for name in characters.keys():
            reply = generate_response(name)
            print(f"{name}: {reply}\n")

            conversation_history.append({"role": "assistant", "content": f"{name}: {reply}"})
```

```python
simulate_conversation()
```
