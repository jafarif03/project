```python
characters = {
    "Researcher": "You are a researcher specialized in climate change. You should explain how global warming is affecting the Earth's temperature, weather patterns, and ecosystems...",
    "Philosopher": "You are a philosopher who analyzes the ethical and philosophical implications of global warming...",
    "Artist": "You are an artist who uses metaphors, visuals, and emotions to express the impact of global warming...",
    "Politician": "You are a politician tasked with finding practical solutions to combat global warming..."
}
```

```python
memory = {name: [] for name in characters}


affinity = {a: {b: 0.0 for b in characters if b != a} for a in characters}
```

```python
conversation_history = [
    {"role": "system", "content": "This is a discussion about global warming between a Researcher, a Philosopher, an Artist, and a Politician. Each participant responds directly to the previous messages, ensuring a logical and engaging conversation."},
    {"role": "user", "content": "Moderator: Let's discuss global warming. What are the biggest challenges we face?"}
]
```

```python
conversation_log = []
affinity_log = []
```

```python
def detect_mentions(message, speaker):
    mentioned = []
    for name in characters:
        if name != speaker and name.lower() in message.lower():
            mentioned.append(name)
    return mentioned
```

```python
def sentiment_toward_target(speaker, target, message):
    prompt = f"""
You are an expert in sentiment analysis. Analyze the sentiment of the following sentence written by "{speaker}" about "{target}". 
Classify it as one of the following: "positive", "negative", or "neutral".
Sentence: "{message}"
Respond with only one word: positive / negative / neutral
"""
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )
    sentiment = response.choices[0].message.content.strip().lower()
    return sentiment
```

```python
def analyze_affinity_change_v2(speaker, message):
    mentions = detect_mentions(message, speaker)
    for target in mentions:
        sentiment = sentiment_toward_target(speaker, target, message)

        if sentiment == "positive":
            affinity[target][speaker] += 0.2
        elif sentiment == "negative":
            affinity[target][speaker] -= 0.2
        elif sentiment == "neutral":
            affinity[target][speaker] += 0.0

        affinity[target][speaker] = max(-1.0, min(1.0, affinity[target][speaker]))
```

```python
def update_memory_with_affinity(listener, speaker, message):
    score = affinity[listener][speaker]
    if score > -0.5:
        memory[listener].append({
            "from": speaker,
            "content": message,
            "affinity": score
        })
```

```python
def get_personal_memory_for_prompt(name, top_k=3):
    relevant = sorted(
        memory[name],
        key=lambda x: x["affinity"],
        reverse=True
    )
    top_memories = relevant[:top_k]
    return [{"role": "user", "content": f'{m["from"]}: {m["content"]}'} for m in top_memories]
```

```python
def generate_response(name):
    recent = conversation_history[-5:]
    messages = recent + [{"role": "system", "content": characters[name]}]
    messages += get_personal_memory_for_prompt(name)

    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=messages
    )
    return response.choices[0].message.content
```

```python
def simulate_conversation(rounds=5):
    for i in range(rounds):
        for speaker in characters.keys():
            reply = generate_response(speaker)
            print(f"{speaker}: {reply}\n")

          
            conversation_log.append({
                "round": i + 1,
                "speaker": speaker,
                "message": reply
            })

            
            conversation_history.append({"role": "assistant", "content": f"{speaker}: {reply}"})

           
            for listener in characters:
                if listener != speaker:
                    update_memory_with_affinity(listener, speaker, reply)

           
            analyze_affinity_change_v2(speaker, reply)

        
        affinity_snapshot = {
            "round": i + 1,
            "affinity": {a: affinity[a].copy() for a in affinity}
        }
        affinity_log.append(affinity_snapshot)

    print("\nüîó Final Affinity Matrix:")
    for a in affinity:
        print(f"{a}: {affinity[a]}")
    save_logs()
```

```python
def save_logs():
    with open("conversation_log.json", "w", encoding="utf-8") as f:
        json.dump(conversation_log, f, indent=2, ensure_ascii=False)

    with open("affinity_log.json", "w", encoding="utf-8") as f:
        json.dump(affinity_log, f, indent=2, ensure_ascii=False)

    print("\nüìÅ Logs saved to 'conversation_log.json' and 'affinity_log.json'")
```

```python
simulate_conversation(rounds=5)
```

```python
import json
import networkx as nx
import matplotlib.pyplot as plt
```

```python
with open("affinity_log.json", "r", encoding="utf-8") as f:
    affinity_log = json.load(f)
```

```python
positions = {
    "Researcher": (-1, 1),
    "Philosopher": (1, 1),
    "Artist": (-1, -1),
    "Politician": (1, -1)
}
```

```python
def edge_color(value):
    if value > 0.1:
        return "green"
    elif value < -0.1:
        return "red"
    else:
        return "gray"
```

```python
for snapshot in affinity_log:
    round_num = snapshot["round"]
    affinities = snapshot["affinity"]
    
    G = nx.DiGraph()
    G.add_nodes_from(affinities.keys())
    
    for source in affinities:
        for target in affinities[source]:
            weight = affinities[source][target]
            G.add_edge(source, target, weight=weight)

    plt.figure(figsize=(6, 5))
    edges = G.edges(data=True)
    weights = [abs(data['weight']) * 5 for _, _, data in edges]
    colors = [edge_color(data['weight']) for _, _, data in edges]

    nx.draw(
        G,
        pos=positions,
        with_labels=True,
        node_color="skyblue",
        edge_color=colors,
        width=weights,
        node_size=2000,
        font_size=10,
        arrows=True,
        arrowstyle='-|>',
        arrowsize=15
    )

    plt.title(f"üîó Affinity Graph - Round {round_num}")
    plt.show()
```
